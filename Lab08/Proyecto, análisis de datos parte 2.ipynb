```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("TEA-Data.csv") 
data
```

```python
data= data.groupby(["Day","#_of_Units_Produced"], sort=False) ["Throughput"].max()
data  
```

```python
ax= data.plot(kind= "bar", subplots= True)
plt.show()   
```

```python
ax= data.plot(kind= "bar", subplots= False)
ax.set_title("Rendimiento por cada unidad producida")
ax.set_xlabel('Día/unidades producidas')
ax.set_ylabel('Rendimiento')
plt.show() 
```

```python
# Gráfica del tiempo en días (Se observa que el rendimiento varía)
data.unstack().T.plot(kind="bar", subplots=False, layout=(2,2),figsize=(20,10), width=2) 
plt.show()  
# Necesito poner titulo al eje y como :Rendimiento
# Y titulo de la gráfica : Número de unidades producidas
```
